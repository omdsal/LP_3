```python
 import time
 def solve_n_queens(n=4, visualize=True, delay=0.25, find_all=False):
    board = [['.' for _ in range(n)] for _ in range(n)]
    cols = set()
    diag1 = set()  # r - c
    diag2 = set()  # r + c
    steps = {'count': 0}
    solutions = []
    def print_board():
        for row in board:
            print(' '.join(row))
        print()
    def step(action, r, c):
        steps['count'] += 1
        print(f"Step {steps['count']:3d}: {action} at (r={r}, c={c})")
        print_board()
        if visualize and delay > 0:
            time.sleep(delay)
    def backtrack(r):
        if r == n:
            print(f"Solution #{len(solutions)+1} found.\n")
            solutions.append([''.join(row) for row in board])
            return
        for c in range(n):
            if c in cols or (r - c) in diag1 or (r + c) in diag2:
                continue
            # place
            cols.add(c); diag1.add(r - c); diag2.add(r + c)
            board[r][c] = 'Q'
            step("Place Q", r, c)
            backtrack(r + 1)
            if not find_all and solutions:
                return
            # remove (backtrack)
            cols.remove(c); diag1.remove(r - c); diag2.remove(r + c)
            board[r][c] = '.'
            step("Remove Q (backtrack)", r, c)
    backtrack(0)
    print(f"Total steps: {steps['count']}, Solutions found: {len(solutions)}")
    return solutions
 if __name__ == "__main__":
    # Example usage:
    # n = 4 (small); set delay smaller/zero to speed up; set find_all=True to enume
    solve_n_queens(n=4, visualize=True, delay=0.2, find_all=False)
```
